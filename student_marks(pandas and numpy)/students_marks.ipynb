```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("students_marks.csv")
```

```python
df
```

```python
df["Total"] = df[["Math", "English", "Science"]].sum(axis=1)
df["Avg"] = df[["Math", "English", "Science"]].mean(axis=1)
```

```python
df
```

```python
df["Result"] = np.where((df[["Math", "English", "Science"]] >=40).all(axis=1), "Pass", "Fail")
```

```python
df
```

```python
topper = df.loc[df["Total"].idxmax()]
print("🏆 Topper:", topper["Name"])
print(topper[["Math", "English", "Science", "Total", "Avg"]])
```
